```python
waterlines = pd.read_csv("waterlines.csv")
berms = pd.read_csv("temp_berms.csv")

waterlines_nd = pd.read_csv("waterlines_ND.csv")
berms_ND =  pd.read_csv("temp_berms_nd.csv")

waterlines_o = pd.read_csv("waterlines_o.csv")
berms_o = pd.read_csv("temp_berms_o.csv")
```

```python
# PLOT CHECK

# Define a colormap
cmap = plt.cm.viridis  # color ramp for throats
cmap2 = plt.cm.Blues # color ramp for waterlines

# Normalize color mapping
num_lines = berms.shape[0]
print(num_lines)
colors = [cmap(i / (num_lines - 1)) for i in range(num_lines)]
colors2 = [cmap2(i / (num_lines - 1)) for i in range(num_lines)]

# Plot each line with a different color
for i, color in enumerate(colors):
    plt.plot(np.arange(berms.shape[1]), 4 - berms.iloc[i, :], color=color, linewidth=1.5)

for i, color in enumerate(colors2):
    plt.plot(np.arange(berms.shape[1]), waterlines.iloc[i, :], color=color, linewidth=1.5)

# plt.plot(np.arange(berms_o.shape[1]), 4 - berms_o.T, ':k', linewidth = 1)
plt.plot(np.arange(berms_ND.shape[1]), 4 - berms_ND.T, ':k', linewidth = 1)


plt.show()
```
