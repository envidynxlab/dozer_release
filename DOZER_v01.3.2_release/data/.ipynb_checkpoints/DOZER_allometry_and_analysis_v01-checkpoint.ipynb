```python
def process_files(file_type1, file_type2):
    # Arrays to store results
    datetime_stamp = []
    trial_id = []
    time = []
    wash_A = []
    wash_P = []
    wash_V = []
    wash_id = []

    # set min threshold for sand binary
    n = 0.01

    # Search for files matching the file_type
    pattern1 = f"*_{file_type1}.csv"
    files1 = glob.glob(pattern1)

    pattern2 = f"*_{file_type2}.csv"
    files2 = glob.glob(pattern2)

    for file1 in files1:
        try:
            # Extract the H/M/S part of the timestamp from the filename
            filename1 = os.path.basename(file1)

            # Extract the full datetime stamp
            dt_stamp1 = filename1.split('_')[0]  # extract datetime stamp part of file name
            
            # Extract just the time stamp
            timestamp1 = filename1.split('_')[0].split('-')[1]  # Extract H/M/S part

            # Extract the trial number
            trial_number1 = re.search(r'trial(\d+)', filename1)
            if trial_number1:
                trial_number1 = trial_number1.group(1)  # Get the number part

            # Load CSV into a numpy array
            data1 = np.genfromtxt(file1, delimiter=',')

            # Create binary mask of sand or empty
            mask_full1 = (data1 > n).astype(int)
            mask1 = mask_full1[3:]

            # Label connected components in the binary mask
            labeled_mask_base = label(mask1)

            for file2 in files2:
                try:
                    # Extract the H/M/S part of the timestamp from the filename
                    filename2 = os.path.basename(file2)

                    # Extract the full datetime stamp
                    dt_stamp2 = filename2.split('_')[0]  # extract datetime stamp part of file name

                    # Extract just the time stamp
                    timestamp2 = filename2.split('_')[0].split('-')[1]  # Extract H/M/S part

                    # Extract the trial number
                    trial_number2 = re.search(r'trial(\d+)', filename2)
                    if trial_number2:
                        trial_number2 = trial_number2.group(1)

                        if trial_number2 == trial_number1:
                            # Load CSV into a numpy array
                            data2 = np.genfromtxt(file2, delimiter=',')

                            # Create binary mask of sand or empty
                            mask_full2 = (data2 > n).astype(int)
                            mask2 = mask_full2[3:]

                            # Label connected components in the binary mask
                            labeled_mask2 = label(mask2)

                            # Get properties of labeled regions
                            regions = regionprops(labeled_mask2)

                            # Iterate over regions to calculate area and perimeter
                            for region in regions:
                                A = region.area  # Number of white pixels
                                P = perimeter(region.image)  # Perimeter of the region

                                # Calculate the volume of the corresponding region within data2[3:]
                                V = np.sum(data2[3:][region.coords[:, 0], region.coords[:, 1]])

                                # Assign label to corresponding "expected" washover deposit
                                eye_dee = mode(labeled_mask_base[region.coords[:, 0], region.coords[:, 1]], axis=None)

                                if A > 0 and P > 30:
                                    # Append results to lists
                                    datetime_stamp.append(dt_stamp2)
                                    trial_id.append(trial_number2)
                                    time.append(timestamp2)
                                    wash_A.append(A)
                                    wash_P.append(P)
                                    wash_V.append(V)
                                    wash_id.append(eye_dee[0])

                except Exception as e:
                    print(f"Error processing file {file2}: {e}")

        except Exception as e:
            print(f"Error processing file {file1}: {e}")

    return datetime_stamp, trial_id, time, wash_id, wash_A, wash_P, wash_V

datetime_stamp, trial, time_D, wash_ID_D, Area_D, Perimeter_D, Volume_D = process_files("final_sand_no_dozer", "temp_sand_dozer")
datetime_stamp, trial, time_ND, wash_ID_ND, Area_ND, Perimeter_ND, Volume_ND = process_files("final_sand_no_dozer", "temp_sand_no_dozer")
```

```python
# Create a DataFrame from the results
df_DOZER = pd.DataFrame({
    'date stamp': datetime_stamp,
    'trial': trial,
    'time': time_D,
    'deposit_id': wash_ID_D
    'perimeter DOZER': Perimeter_D,
    'area DOZER': Area_D,
    'volume DOZER': Volume_D
})

df_DOZER["time"] = df_DOZER.time.astype(float)
df_DOZER.sort_values(by=['time'], inplace=True)
```
