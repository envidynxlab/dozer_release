## The purpose of this notebook is to download and organise datasets presented in Lazarus et al. (DOI), and generate the plots shown in that manuscript.


* TES data
* Lazarus et al. (Earth's Future, 2021: https://doi.org/10.1029/2020EF001818) – data via Taylor et al. (figshare, 2021: https://doi.org/10.6084/m9.figshare.12608828.v1)
* Lazarus (2016: https://doi.org/10.1002/2016GL071213) – data via figshare: https://doi.org/10.5281/zenodo.6337518

```python
# Import libraries

import pandas as pd
import numpy as np
import math

# Import curve-fitting package from scipy
from scipy.optimize import curve_fit
from scipy import stats

import matplotlib as mpl
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

## (1) Downloading & organising the TES dataset:

```python
# Import the morphometric measurements for the TES experiments:
TES = pd.read_csv("TES_washover_morphometry.csv")

TES.head(5) # check to confirm
```

```python
# To calcuate the distortion index (DI) based on measured perimeter & deposit area...

# Pull the necessary series:
Pmx = TES.loc[:,'deposit_perimeter_m'] # perimeter measured
A = TES.loc[:,'deposit_area_m2'] # deposit area

# Calculate idealised perimeter of a closed semi-circle, given conservation of deposit area:
Pi = (A
      .mul(2)
      .div(math.pi)
      .pow(0.5)
      .mul((2 + math.pi))
     )

# Calculate the distortion index (DI):
DI = (Pmx
      .div(Pi)
     )

# Add idealised perimeter and distortion index to the dataframe
TES['perimeter_ideal'] = Pi
TES['distortion_index'] = DI

TES.head(3) # check to confirm
```

```python
# Reorganise column sequence for tidiness
cols = ['id',
        'throat_width_m',
        'deposit_width_m',
        'intrusion_length_m',
        'deposit_area_m2',
        'deposit_volume_m3',
        'deposit_perimeter_m',
        'perimeter_ideal',
        'distortion_index',
        'built_fraction',
        'forcing_mode',
        'dataset']

TES = TES[cols]

TES.head(3) # check to confirm 
```

```python
# Sort dataframe by built fraction, forcing mode, and dataset:

TES.sort_values(by=['built_fraction', 'forcing_mode'], inplace=True)

TES.tail(5) # check bottom of dataframe to confirm 
```

```python
# Round (bin) built fraction to nearest tenth, and add that rounded value to the dataframe:

# pull necessary series:
bf = TES.loc[:,'built_fraction'] # built fraction

bf_round = round(bf, 1) # round elements in series to one decimal place
bf_round.tail(10) # check to confirm 
```

```python
# Add rounded built fraction to the dataframe, and rearrange the columns

TES.insert(10, 'bf_rounded', bf_round)

TES.tail(10) # check to confirm 
```

**A few summary (aggregate) statistics:**

```python
# how many deposits from each forcing mode?

fm = pd.DataFrame(TES
       .groupby('forcing_mode')
       .agg('count')
       .loc[:, 'id']
      )

fm.rename({'id': 'counts'}, axis=1, inplace=True)
fm.reset_index(inplace=True)

fm
```

```python
# how many deposits in each condition for (rounded) built fraction?

bfs = pd.DataFrame(TES
       .groupby('bf_rounded')
       .agg('count')
       .loc[:, 'id']
      )

bfs.rename({'id': 'counts'}, axis=1, inplace=True)
bfs.reset_index(inplace=True)

bfs
```

```python
# Compile various counts from TES experiments:

TES_counts = pd.DataFrame({'subset':
                        ['non-built',
                         'built',
                         'inundation',
                         'waves',
                         'binned bf = 0',
                         'binned bf = 0.2',
                         'binned bf = 0.3',
                         'binned bf = 0.4',
                         'binned bf = 0.5'],
                           'counts':np.zeros(9)})

TES_counts['counts']=TES_counts['counts'].astype(int)


# populate the table

TES_counts.iloc[0,1] = bfs.iloc[0,1] # total non-built

TES_counts.iloc[1,1]=bfs.iloc[1:,1].sum() # total built

TES_counts.iloc[2:4,1]=fm.iloc[:, 1].values # forcing mode

TES_counts.iloc[4:,1]=bfs.iloc[:,1].values # various built fractions

TES_counts.set_index('subset', inplace=True) # change the index (for clarity)

TES_counts
```

## (2) Downloading & organising the Taylor et al. (2021) dataset: https://doi.org/10.6084/m9.figshare.12608828.v1

```python
# to pull the 'built' data

url1 = 'https://figshare.com/ndownloader/files/23820212'
TB = pd.read_csv(url1) # raw dataframe for the Taylor 'built' data

TB # check to confirm 
```

```python
# get rid of last row (empties)...
TB = TB.drop(TB.index[-1])
TB.tail(5)
```

```python
# Rename columns for tidiness...
TB.rename({'OBJECTID': 'id',
           'length': 'intrusion_length_m',
           'area': 'deposit_area_m2',
           'perimeter': 'deposit_perimeter_m',
           'ideal_perimeter': 'perimeter_ideal',
           'building_fraction': 'built_fraction'
             }, axis=1, inplace=True)


# Create a new categorial column for mode of forcing (here, 'inundation') & add to dataframe
forcing = pd.Series(['mixed']*(TB.shape[0]), dtype='category')
TB['forcing_mode'] = forcing

# Create a new categorical column for dataset (to distinguish from others)
dataset = pd.Series(['Taylor']*(TB.shape[0]), dtype='category')
TB['dataset'] = dataset


# Reorganise column sequence for tidiness
cols = ['id',
        'intrusion_length_m',
        'deposit_area_m2',
        'deposit_perimeter_m',
        'perimeter_ideal',
        'distortion_index',
        'built_fraction',
        'forcing_mode',
        'dataset',
        'hurricane']

TB = TB[cols]


# sort by 'built fraction' (low to high)
TB.sort_values(by=['built_fraction'], inplace=True)

TB.head(10) # check to confirm
```

```python
# to pull the 'non-built' data
url2 = 'https://figshare.com/ndownloader/files/23820266' 
TnB = pd.read_csv(url2) # raw dataframe for the Taylor 'non-built' data

TnB.head(5) # check to confirm 
```

```python
# Rename columns for tidiness...
TnB.rename({'OBJECTID': 'id',
           'length': 'intrusion_length_m',
           'area': 'deposit_area_m2',
           'perimeter': 'deposit_perimeter_m',
           'ideal_perimeter': 'perimeter_ideal',
             }, axis=1, inplace=True)


# Create a new categorial column for mode of forcing (here, 'inundation') & add to dataframe
forcing = pd.Series(['mixed']*(TnB.shape[0]), dtype='category')
TnB['forcing_mode'] = forcing

# Create a new categorical column for dataset (to distinguish from others)
dataset = pd.Series(['Taylor']*(TnB.shape[0]), dtype='category')
TnB['dataset'] = dataset

# Note that these are deposits in 'non-built' settings, so built fraction == 0
bf_zero = pd.Series(np.zeros(TnB.shape[0]), dtype='float')
TnB['built_fraction'] = bf_zero


# Reorganise column sequence for tidiness
cols = ['id',
        'intrusion_length_m',
        'deposit_area_m2',
        'deposit_perimeter_m',
        'perimeter_ideal',
        'distortion_index',
        'built_fraction',
        'forcing_mode',
        'dataset',
        'hurricane']

TnB = TnB[cols]


TnB.head(10) # check to confirm
```

```python
# Join the two 'Taylor' dataframes into a single dataframe that combines the Taylor built & non-built sets:

Taylor_compiled = pd.concat([TnB, TB], ignore_index=True, sort=False)

Taylor_compiled.head(5)
```

```python
# Round (bin) built fraction to nearest tenth, and add that rounded value to the dataframe:

# pull necessary series:
Tbf = Taylor_compiled.loc[:,'built_fraction'] # built fraction

Tbf_round = round(Tbf, 1) # round elements in series to one decimal place


# Add rounded built fraction to the dataframe, and rearrange the columns
Taylor_compiled.insert(7, 'bf_rounded', Tbf_round)

Taylor_compiled.tail(10) # check to confirm 
```

## (3) Downloading & organising the SAFL experiments dataset from Lazarus (2016): https://doi.org/10.1002/2016GL071213

```python
url = 'https://zenodo.org/record/6337518/files/Lazarus_GRL2016_SAFLdata_WASHOVER.csv?download=1'

SAFL = pd.read_csv(url) # raw dataframe for the Taylor 'non-built' data

SAFL.head(5) # check to confirm 
```

**Join all three 'dataset' dataframes into a single dataframe of all available washover data:**

```python
# Join all three 'dataset' dataframes into a single dataframe of all available washover data:
WASHOVER = pd.concat([TES, Taylor_compiled, SAFL], ignore_index=True, sort=False)

WASHOVER.head(3)
```

```python
# Drop incomplete columns that aren't relevant here – so these:
WASHOVER = WASHOVER.drop(labels=['id', 'throat_width_m', 'hurricane', 'trial'], axis=1)

WASHOVER.head(3)

# Data are now ready for plotting...
```

```python
# Fits & KS test for inundation vs waves

TES = TES.rename(columns={'forcing mode': 'forcing_mode'})

# new dataframe for each forcing mode:
IND = TES[(TES.forcing_mode == 'inundation')]
WAV = TES[(TES.forcing_mode == 'waves')]

# pull V & A from each subset
V_ind = IND.loc[:,'deposit_volume_m3']
A_ind = IND.loc[:,'deposit_area_m2']

V_wav = WAV.loc[:,'deposit_volume_m3']
A_wav = WAV.loc[:,'deposit_area_m2']


# calc fits

# Function to calculate the power-law with constants a and b
def power_law(x, a, b):
    return a*np.power(x, b)

# INUNDATION:

# Fit the data
pars_ind, cov_ind = curve_fit(f=power_law, xdata=A_ind, ydata=V_ind, p0=[0, 0], bounds=(-np.inf, np.inf))


# Get the standard deviations of the parameters (square roots of the # diagonal of the covariance)
stdevs_ind = np.sqrt(np.diag(cov_ind))

# Calculate the residuals:
res_ind = V_ind - power_law(A_ind, *pars_ind)

# Calculate the residuals sum of squares:
ss_res_ind = np.sum(res_ind**2)

# Total sum of squares (ss_tot):
ss_tot_ind = np.sum((V_ind - np.mean(V_ind))**2)

# R_squared-value:
R_squared_ind = 1 - (ss_res_ind / ss_tot_ind)


# WAVES:

# Fit the data
pars_wav, cov_wav = curve_fit(f=power_law, xdata=A_wav, ydata=V_wav, p0=[0, 0], bounds=(-np.inf, np.inf))


# Get the standard deviations of the parameters (square roots of the # diagonal of the covariance)
stdevs_wav = np.sqrt(np.diag(cov_wav))

# Calculate the residuals:
res_wav = V_wav - power_law(A_wav, *pars_wav)

# Calculate the residuals sum of squares:
ss_res_wav = np.sum(res_wav**2)

# Total sum of squares (ss_tot):
ss_tot_wav = np.sum((V_wav - np.mean(V_wav))**2)

# R_squared-value:
R_squared_wav = 1 - (ss_res_wav / ss_tot_wav)
```

```python
# Import the morphometric measurements for the Sandy measurements (New Jersey):
NJ = pd.read_csv('NJ_Sandy_metrics.csv')

NJ.head(5) # check to confirm
```

```python
# To calcuate the distortion index (DI) based on measured perimeter & deposit area...

# Pull the necessary series:
Pmx = NJ.loc[:,'deposit_perimeter_m'] # perimeter measured
A = NJ.loc[:,'deposit_area_m2'] # deposit area

# Calculate idealised perimeter of a closed semi-circle, given conservation of deposit area:
Pi = (A
      .mul(2)
      .div(math.pi)
      .pow(0.5)
      .mul((2 + math.pi))
     )

# Calculate the distortion index (DI):
DI = (Pmx
      .div(Pi)
     )

# Add idealised perimeter and distortion index to the dataframe
NJ['perimeter_ideal'] = Pi
NJ['distortion_index'] = DI

NJ.head(3) # check to confirm
```

```python
# Sort dataframe by built fraction, forcing mode, and dataset:

NJ.sort_values(by=['built_fraction'], inplace=True)

NJ.tail(5) # check bottom of dataframe to confirm 
```

```python
# Round (bin) built fraction to nearest tenth, and add that rounded value to the dataframe:

# pull necessary series:
bf = NJ.loc[:,'built_fraction'] # built fraction

bf_round = round(bf, 1) # round elements in series to one decimal place
# bf_round.tail(10) # check to confirm

NJ['bf_rounded'] = bf_round

NJ.head(3)
```

```python
# Trying to curve fit...

# Function to calculate the power-law with constants a and b
def power_law(x, a, b):
    return a*np.power(x, b)


# Data to fit
A = ALLVOLS.loc[:,'deposit_area_m2'] # deposit area
V = ALLVOLS.loc[:,'deposit_volume_m3'] # perimeter measured


# Fit the data
parsAV, cov = curve_fit(f=power_law, xdata=A, ydata=V, p0=[0, 0], bounds=(-np.inf, np.inf))

# Get the standard deviations of the parameters (square roots of the # diagonal of the covariance)
stdevsAV = np.sqrt(np.diag(cov))



# Calculate the residuals:
res = V - power_law(A, *parsAV)

# Calculate the residuals sum of squares:
ss_res = np.sum(res**2)

# Total sum of squares (ss_tot):
ss_tot = np.sum((V - np.mean(V))**2)

# R_squared-value:
R_squared_AV = 1 - (ss_res / ss_tot)

```
