```python
# Import libraries

import pandas as pd
import numpy as np
import math

import os
import glob
import re

from datetime import datetime

from scipy.stats import mode

import matplotlib as mpl
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
from matplotlib.collections import LineCollection
import matplotlib.colors as mcolors
from matplotlib.colors import Normalize, LinearSegmentedColormap

import matplotlib.pyplot as plt
import seaborn as sns

from skimage.measure import label, regionprops, perimeter

%matplotlib inline
```

```python
def process_files(file_type1, file_type2, file_type3):
    # Arrays to store results
    datetime_stamp = []
    trial_id = []
    time = []
    wash_A = []
    wash_P = []
    wash_V = []
    wash_id = []

    # set min threshold for sand binary
    sand_n = 0.02
    ow_n = 0
    row = 1

    # Search for files matching the file_type
    pattern1 = f"*_{file_type1}.csv"
    files1 = glob.glob(pattern1)

    # pattern2 = f"*_{file_type2}.csv"
    # files2 = glob.glob(pattern2)

    def extract_timestamp(filename):
        """Extract timestamp from filename assuming it follows 'YYYYMMDD-HHMMSS' format at the start."""
        match = re.match(r"(\d{8}-\d{6})", filename)  # Adjust regex based on actual format
        if match:
            return datetime.strptime(match.group(1), "%Y%m%d-%H%M%S")
        return None

    # Search for file_type2 files
    pattern2 = f"*_{file_type2}.csv"
    files2 = glob.glob(pattern2)
    
    # Sort files2 based on extracted timestamp
    files2 = sorted(files2, key=lambda f: extract_timestamp(os.path.basename(f)))

    
    pattern3 = f"*_{file_type3}.csv"
    files3 = glob.glob(pattern3)

    # File 1 is the final ND condition – what we'd expect to see for these conditions
    for file1 in files1:
        try:
            # Extract the H/M/S part of the timestamp from the filename
            filename1 = os.path.basename(file1)

            # Extract the full datetime stamp
            dt_stamp1 = filename1.split('_')[0]  # extract datetime stamp part of file name

            # Extract just the time stamp
            timestamp1 = filename1.split('_')[0].split('-')[1]  # Extract H/M/S part

            # Extract the trial number
            trial_number1 = re.search(r'trial(\d+)', filename1)
            if trial_number1:
                trial_number1 = trial_number1.group(1)  # Get the number part

            # Load CSV into a numpy array
            data1 = np.genfromtxt(file1, delimiter=',')

            # Create binary mask of sand or empty
            mask_full1 = (data1 > sand_n).astype(int)
            mask1 = mask_full1[row:]

            
            # Label connected components in the binary mask
            labeled_mask_base = label(mask1)
            
            unique_values = np.unique(labeled_mask_base)
            unique_values = unique_values[unique_values != 0]  # Remove zero
            
            # print("Unique values: ", len(unique_values))

            store_ow_mask = np.zeros((mask1.shape[0], mask1.shape[1], len(unique_values)))
            cumulative_ow_mask = np.zeros((mask1.shape[0], mask1.shape[1]))
            # print("got here")

            # plt.imshow(labeled_mask_base, cmap = 'Reds')
            # plt.pause(0.5)  # Pause for 0.5 seconds to see each frame
            # plt.colorbar()
            # plt.clf()

            # print('Got here...')

            # print("Max labels: ", labeled_mask_base.max())
            
            # File 2 is the overwash footprint at each pulse
            for file2 in files2:
                try:
                    # Extract the H/M/S part of the timestamp from the filename
                    filename2 = os.path.basename(file2)

                    # Extract the full datetime stamp
                    dt_stamp2 = filename2.split('_')[0]  # extract datetime stamp part of file name

                    # Extract just the time stamp
                    timestamp2 = filename2.split('_')[0].split('-')[1]  # Extract H/M/S part

                    # Extract the trial number
                    trial_number2 = re.search(r'trial(\d+)', filename2)
                    if trial_number2:
                        trial_number2 = trial_number2.group(1)

                        # Match the trial numbers
                        if trial_number2 == trial_number1:
                            # Load CSV into a numpy array
                            data2 = np.genfromtxt(file2, delimiter=',')

                            # File 3 is the underlying sand cover at time of pulse
                            for file3 in files3:
                                try:
                                    # Extract the H/M/S part of the timestamp from the filename
                                    filename3 = os.path.basename(file3)
                
                                    # Extract the full datetime stamp
                                    dt_stamp3 = filename3.split('_')[0]  # extract datetime stamp part of file name
                
                                    # Extract just the time stamp
                                    timestamp3 = filename3.split('_')[0].split('-')[1]  # Extract H/M/S part
                
                                    # Extract the trial number
                                    trial_number3 = re.search(r'trial(\d+)', filename3)
                                    if trial_number3:
                                        trial_number3 = trial_number3.group(1)
                
                                        if trial_number3 == trial_number1:

                                            if timestamp3 == timestamp2:
                                                # Load CSV into a numpy array
                                                data3 = np.genfromtxt(file3, delimiter=',')
    
                                                # Create binary mask of sand or empty
                                                mask_full2 = (data2 > ow_n).astype(int)
                                                mask2 = mask_full2[row:]
                    
                                                # Label connected components in the binary mask
                                                labeled_mask2 = label(mask2)
                                                
                                                tlm = labeled_mask2.copy()
                                                tlm[tlm > 0] = 1
                                                cumulative_ow_mask += tlm

                                                # Inside your loop:
                                                # plt.imshow(cumulative_ow_mask, cmap='Blues')
                                                # plt.title(f"Trial: {trial_number3}, Ts: {timestamp3}")  # Label each iteration
                                                # plt.colorbar()
                                                # plt.pause(0.5)  # Pause for 0.5 seconds to see each frame
                                                # plt.clf()  # Clear figure for the next iteration

                                                # print('labled_mask2', labeled_mask2)
                    
                                                # Get properties of labeled regions
                                                regions = regionprops(labeled_mask2)
                                                
                                                # Inside your loop:
                                                # plt.imshow(labeled_mask2, cmap='Purples')
                                                # plt.title(f"Trial: {trial_number3}, Ts: {timestamp3}")  # Label each iteration
                                                # plt.colorbar()
                                                # plt.pause(0.5)  # Pause for 0.5 seconds to see each frame
                                                # plt.clf()  # Clear figure for the next iteration

                                                # print("num regions: ", len(regions))
                    
                                                # Iterate over regions to calculate area and perimeter
                                                # reg_num = 1
                                                for region in regions:
                                                    # print("Region: ", reg_num)
                                                    # reg_num += 1
                    
                                                    
                                                    # A = region.area  # Number of white pixels
                                                    # P = perimeter(region.image)  # Perimeter of the region
                    
                                                    # data3 = np.genfromtxt(file3, delimiter=',')
                    
                                                    # # Calculate the volume of the corresponding region within data2[3:]
                                                    # V = np.sum(data3[row:][region.coords[:, 0], region.coords[:, 1]])

                                                    # print("Region: ", reg_num)
                                                    # print("A: ", A)
                                                    # print("P: ", P)
                                                    # print("V: ", V)
                                                    # reg_num += 1
                                                    
                    
                                                    # # Assign label to corresponding "expected" washover deposit
                                                    # eye_dee = mode(labeled_mask_base[region.coords[:, 0], region.coords[:, 1]], axis=None)
                    
                                                    # Extract the values of labeled_mask_base at the region coordinates
                                                    values = labeled_mask_base[region.coords[:, 0], region.coords[:, 1]]
                                                    
                                                    # values = labeled_mask2[region.coords[:, 0], region.coords[:, 1]]
                                                    # print("values: ", values)
                                                    
                                                    # Remove zeros
                                                    non_zero_values = values[values > 0]
                                                    
                                                    if non_zero_values.size > 0:
                                                        eye_dee = mode(non_zero_values, axis=None)[0]
                                                        
                                                        # store_ow_mask = np.zeros((mask1.shape[0], mask1.shape[1], len(unique_values)))
                                                        # store_ow_mask[region.coords[:, 0], region.coords[:, 1], (eye_dee-1)] = eye_dee
                                                        
                                                        store_ow_mask[region.coords[:, 0], region.coords[:, 1], (eye_dee-1)] += 1


                                                        ## Collect data:

                                                        # Create a blank 2D mask of the same size as 'store_ow_mask' (first two dimensions)
                                                        temp_mask = np.zeros(store_ow_mask.shape[:2], dtype=int)
                                                        
                                                        # Set the relevant region's pixels
                                                        ####### Don't want the region coords – want the MASK coords
                                                        temp_mask[store_ow_mask[:, :, (eye_dee-1)] > 0] = 1  # Mark region as 1 (binary mask)
                                                        # plt.imshow(temp_mask, cmap = 'Reds')

                                                        # Inside your loop:
                                                        # plt.imshow(temp_mask, cmap='Reds')
                                                        # plt.title(f"Iteration {eye_dee}")  # Label each iteration
                                                        # # plt.colorbar()
                                                        # plt.pause(0.2)  # Pause for 0.5 seconds to see each frame
                                                        # plt.clf()  # Clear figure for the next iteration

                                                        # Label the regions
                                                        temp_mask_labeled = label(temp_mask)
                                                        
                                                        # Extract region properties
                                                        temp_reg = regionprops(temp_mask_labeled)

                                                        
                                                        # # A = temp_reg[0].area  # Number of white pixels
                                                        # A = temp_mask.sum()  # Number of white pixels
                                                        # P = perimeter(temp_reg[0].image)  # Perimeter of the region
                        
                                                        data3_full = np.genfromtxt(file3, delimiter=',')
                                                        data3 = data3_full[row:]


                                                        sand_select = data3.copy()
                                                        sand_select[temp_mask < 1] = 0
                                                        sand_select[sand_select < sand_n] = 0

                                                        select_mask = sand_select.copy()
                                                        select_mask[select_mask > 0] = 1

                                                        if select_mask.sum().sum() > 0:
                                                            select_mask_labeled = label(select_mask)
                                                            select_reg = regionprops(select_mask_labeled)

                                                            # print("select_reg: ", select_reg[0])
    
                                                            # A = select_reg[0].area
                                                            # A = select_mask.sum().sum()
                                                            A = sum(region.area for region in select_reg)
                                                            # P = perimeter(select_reg[0].image)
                                                            P = sum(perimeter(region.image) for region in select_reg)
                                                            V = np.sum(sand_select)
    
                                                            # print("A: ", A)
                                                            # print("P: ", P)
                                                            
                                                            # print("got here")

                                                            # plt.imshow(sand_select, cmap='Reds')
                                                            # plt.title(f"Trial: {trial_number3}, Ts: {timestamp3}, Deposit_id: {eye_dee}")  # Label each iteration
                                                            # plt.colorbar()
                                                            # plt.pause(0.1)  # Pause for 0.5 seconds to see each frame
                                                            # plt.clf()  # Clear figure for the next iteration

                                                            

                                                        
                        
                                                        # Calculate the volume of the corresponding region within data2[3:]
                                                        # V = np.sum(data3[row:][temp_reg[0].coords[:, 0], temp_reg[0].coords[:, 1]])
                                                        # V = np.sum(data3[temp_mask > 0])
                                                            # V = np.sum(sand_select)

                                                            if A > 4:
                                                                # Append results to lists
                                                                datetime_stamp.append(dt_stamp2)
                                                                trial_id.append(trial_number3)
                                                                time.append(timestamp2)
                                                                wash_A.append(A)
                                                                wash_P.append(P)
                                                                wash_V.append(V)
                                                                wash_id.append(eye_dee)


                                                    
                                                        # print("got to eye dee")

                                                        # print("deposit id: ", eye_dee)
                                                    else:
                                                        eye_dee = 0  # Default value if no non-zero labels exist
                                                        # print("Region: ", reg_num)
                                                        # plt.imshow(labeled_mask_base+labeled_mask2, cmap = 'Reds')
                                                        # # plt.imshow(labeled_mask2, cmap = 'Blues')
                                                        
                    
                                                    # if A > 0 and P > 40:
                                                    # if A > 4:
                                                    #     # Append results to lists
                                                    #     datetime_stamp.append(dt_stamp2)
                                                    #     trial_id.append(trial_number3)
                                                    #     time.append(timestamp2)
                                                    #     wash_A.append(A)
                                                    #     wash_P.append(P)
                                                    #     wash_V.append(V)
                                                    #     wash_id.append(eye_dee)
                                                    #     # wash_id.append(eye_dee)


                                
                                                    # reg_num += 1

                
                                except Exception as e:
                                    print(f"Error processing file {file3}: {e}")

                except Exception as e:
                    print(f"Error processing file {file2}: {e}")
    
        except Exception as e:
            print(f"Error processing file {file1}: {e}")

    return datetime_stamp, trial_id, time, wash_id, wash_A, wash_P, wash_V #, store_ow_mask, cumulative_ow_mask


datetime_stamp, trial_D, time_D, wash_ID_D, Area_D, Perimeter_D, Volume_D = process_files("final_sand_ND", "ow_footprint_D", "ow_sand_D")
datetime_stamp, trial_ND, time_ND, wash_ID_ND, Area_ND, Perimeter_ND, Volume_ND = process_files("final_sand_ND", "ow_footprint_ND", "ow_sand_ND")



# Create a DataFrame from the results
df_DOZER = pd.DataFrame({
    'trial': trial_D,
    'time': time_D,
    'area D': Area_D,
    'perimeter D': Perimeter_D,
    'volume D': Volume_D,
    'deposit_id': wash_ID_D
})

df_DOZER["time"] = df_DOZER.time.astype(float)
df_DOZER.sort_values(by=['time', 'deposit_id'], inplace=True)
df_DOZER.reset_index(drop=True, inplace=True)

df_DOZER = df_DOZER[df_DOZER['deposit_id'] != 0]
df_DOZER = df_DOZER[df_DOZER['volume D'] > 0]



df_NODOZ = pd.DataFrame({
    'trial': trial_ND,
    'time': time_ND,
    'area ND': Area_ND,
    'perimeter ND': Perimeter_ND,
    'volume ND': Volume_ND,
    'deposit_id': wash_ID_ND
})

df_NODOZ["time"] = df_NODOZ.time.astype(float)
df_NODOZ.sort_values(by=['time', 'deposit_id'], inplace=True)
df_NODOZ.reset_index(drop=True, inplace=True)

df_NODOZ = df_NODOZ[df_NODOZ['deposit_id'] != 0]
df_NODOZ = df_NODOZ[df_NODOZ['volume ND'] > 0]


# Function to normalize time within each trial
def normalize_time(group):
    return (group - group.min()) / (group.max() - group.min())

# Apply the normalization
df_DOZER['normalized_time'] = df_DOZER.groupby('trial')['time'].transform(normalize_time)
df_NODOZ['normalized_time'] = df_NODOZ.groupby('trial')['time'].transform(normalize_time)


# Print list of trials
df_NODOZ['trial'].unique()
```

```python
f, ax = plt.subplots(figsize = (6, 6))

custom = {'font': 'Arial',
               'xtick.direction': 'in',
               'ytick.direction': 'in',
               'xtick.top': 'True',
               'ytick.right': 'True'}

sns.set_style('ticks', rc = custom)

# Get the 'Blues' colormap
cmap1 = plt.get_cmap('Blues')
# Get the last (darkest) color in RGB
darkest_rgb1 = cmap1(1.0)

# Get the 'Reds' colormap
cmap2 = plt.get_cmap('Reds')
# Get the last (darkest) color in RGB
darkest_rgb2 = cmap2(1.0)


# Iterate over each trial group
for trial, trial_data in df_merged.groupby(['trial_DOZER', 'deposit_id']):
    # Extract relevant data for the current trial
    x_D = trial_data['area D'].values
    y_D = trial_data['volume D'].values  # Adjusted y
    run_time = trial_data['normalized_time_DOZER'].values  # Color values
    
    x_ND = trial_data['area ND'].values
    y_ND = trial_data['volume ND'].values  # Adjusted y
    
    # Create line segments
    points_D = np.column_stack([x_D, y_D])
    segments_D = np.stack([points_D[:-1], points_D[1:]], axis=1)

    points_ND = np.column_stack([x_ND, y_ND])
    segments_ND = np.stack([points_ND[:-1], points_ND[1:]], axis=1)

    # Create LineCollections with color mapping
    lc_ND = LineCollection(segments_ND, cmap='Blues', norm=plt.Normalize(run_time.min(), run_time.max()))
    lc_ND.set_array(run_time)

    lc_D = LineCollection(segments_D, cmap='Reds', norm=plt.Normalize(run_time.min(), run_time.max()))
    lc_D.set_array(run_time)

    # Plot the line collection for this trial
    ax.add_collection(lc_ND)
    ax.add_collection(lc_D)

    # Markers for start and end points
    ax.plot(x_ND[0], y_ND[0], marker='o', markerfacecolor='white', markeredgecolor='black', markeredgewidth=1, markersize=8)
    ax.plot(x_ND[-1], y_ND[-1], marker='o', markerfacecolor=darkest_rgb1, markeredgecolor='white', markeredgewidth=1, markersize=9)

    ax.plot(x_D[0], y_D[0], marker='o', markerfacecolor='white', markeredgecolor='black', markeredgewidth=1, markersize=8)
    ax.plot(x_D[-1], y_D[-1], marker='o', markerfacecolor=darkest_rgb2, markeredgecolor='white', markeredgewidth=1, markersize=9)

# Adjust plot settings
# ax.autoscale()
# plt.ylabel('mean crest fraction', fontweight='semibold')
# plt.xlabel('total washover area ($\mathregular{L^2}$)', fontweight='semibold')

sns.despine(right=False, top=False)
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_box_aspect(1)

ax.set_xlabel('log [area ($\mathregular{L^2}$)]', fontweight='semibold')
ax.set_ylabel('log [volume ($\mathregular{L^3}$)]', fontweight='semibold')

# Show plot
# plt.show()

plt.tight_layout()  # Adjusts spacing automatically


# plt.savefig('DOZER_dynamic_allom_VA.png', dpi = 300) # saves your plot as a png
```

```python
# Compile and append Qmove into one superfile:

def process_Qmove_files(file_type1):
    
# Search for files matching the file_type
    pattern1 = f"*_{file_type1}.csv"
    files1 = glob.glob(pattern1)

    # Initialize an empty list to store DataFrames
    dataframes = []
    
    # Loop through each CSV file, read it into a DataFrame, and append it to the list
    for file in files1:

        filename1 = os.path.basename(file)

        # Extract the trial number
        trial_number1 = re.search(r'trial(\d+)', filename1)
        if trial_number1:
            trial_number1 = trial_number1.group(1)  # Get the number part
        
        df = pd.read_csv(file)  # Read CSV into DataFrame
        df['trial'] = trial_number1
        
        dataframes.append(df)  # Append to list
    
    # Concatenate all DataFrames into a single DataFrame
    Qmove_data = pd.concat(dataframes, ignore_index=True)

    return Qmove_data


Qmove_data = process_Qmove_files("Qmove_series")

# Function to normalize time within each trial
def normalize_time(group):
    return (group - group.min()) / (group.max() - group.min())

# Apply the normalization
Qmove_data['normalized_time'] = Qmove_data.groupby('trial')['pulse_time'].transform(normalize_time)

Qmove_data['wet_D_diff'] = Qmove_data.groupby('trial')['wet_D'].diff()
Qmove_data['wet_ND_diff'] = Qmove_data.groupby('trial')['wet_ND'].diff()

Qmove_data.head(10)

# Crest data:

# Waterlines:
```

```python

# Copy your original data
df = Qmove_data.copy()

# Round pulse_time
df['pulse_time_rounded'] = df['pulse_time'].round().astype(int)

# Define a function to apply to each group
def process_group(group):
    # Create full time range for this trial
    time_range = np.arange(0, group['pulse_time_rounded'].max() + 1)
    new_df = pd.DataFrame({'pulse_time_new': time_range})
    
    # Merge and fill zeros
    merged = new_df.merge(group[['pulse_time_rounded', 'wet_D', 'wet_ND']],
                          how='left',
                          left_on='pulse_time_new',
                          right_on='pulse_time_rounded')
    
    merged['wet_D'] = merged['wet_D'].fillna(0)
    merged['wet_ND'] = merged['wet_ND'].fillna(0)
    merged = merged.drop(columns='pulse_time_rounded')
    
    # Add trial column back in
    merged['trial'] = group['trial'].iloc[0]
    
    return merged

# Apply the function to each group
result = df.groupby('trial').apply(process_group).reset_index(drop=True)

# Function to normalize time within each trial
def normalize_time(group):
    return (group - group.min()) / (group.max() - group.min())

# Apply the normalization
result['normalized_time_new'] = result.groupby('trial')['pulse_time_new'].transform(normalize_time)

# Now 'result' contains the zero-filled pulse_time_new, wet_D, wet_ND per trial
result.head()
```

```python
# import numpy as np
# import pandas as pd
# import matplotlib.pyplot as plt
# import seaborn as sns
# from matplotlib.collections import LineCollection
# from matplotlib.colors import LinearSegmentedColormap

# Create custom colormap
cmap = LinearSegmentedColormap.from_list("custom_gray", [(0.9, 0.9, 0.9), (0, 0, 0)])
# cmap = 'Reds'


# Set custom plot style
custom = {'font': 'Arial',
          'xtick.direction': 'in',
          'ytick.direction': 'in',
          'xtick.top': True,
          'ytick.right': True}
sns.set_style('ticks', rc=custom)

# # Normalize color mapping globally across all trials
# norm = plt.Normalize(game_data["run_time"].min(), game_data["run_time"].max())

# Custom function to plot each trajectory in a facet
def plot_trajectory(data, color, **kwargs):
    ax = plt.gca()  # Get current axis
    
    x = data["dozer_x"].values
    y = 720 - data["dozer_y"].values  # Corrected transformation
    run_time = data["run_time"].values

    # Create line segments
    points = np.column_stack([x, y])
    segments = np.stack([points[:-1], points[1:]], axis=1)

    # Compute normalization per trial
    trial_norm = plt.Normalize(run_time.min(), run_time.max())

    # Create a LineCollection
    lc = LineCollection(segments, cmap = cmap, norm = trial_norm, linewidth = 2)
    lc.set_array(run_time)
    
    ax.add_collection(lc)  # Add the colored line collection

    # Markers for start and end points
    ax.plot(x[0], y[0], marker='o', markerfacecolor='white', markeredgecolor='black', markeredgewidth=1, markersize=10, zorder=2)
    ax.plot(x[-1], y[-1], marker='o', markerfacecolor='black', markeredgecolor='white', markeredgewidth=1, markersize=11, zorder=2)

    ax.autoscale()  # Scale the axes properly

# Initialize FacetGrid
# grid = sns.FacetGrid(game_data, col="trial", col_wrap=5, height=3, sharex=True, sharey=True, despine=False, aspect=1)
grid = sns.FacetGrid(game_data, 
                     col="trial", 
                     col_wrap=5, 
                     height=6,  # Controls size
                     aspect=1,  # Ensures square plots
                     sharex=True, 
                     sharey=True, 
                     despine=False)

# Map the custom function
grid.map_dataframe(plot_trajectory)

# Set labels and titles
grid.set_axis_labels('Alongshore Position', 'Cross-Shore Position', size = 12, fontweight='semibold')
grid.set_titles(template="Trial: {col_name}", size = 12, fontweight='semibold')

# plt.show()
plt.tight_layout()  # Adjusts spacing automatically
# plt.savefig('DOZER_tracks_trajectories.png', dpi = 300) # saves your plot as a png
```
